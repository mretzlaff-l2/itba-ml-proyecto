```python
# imports
import pandas as pd
import matplotlib.pyplot as plt

# this allows plots to appear directly in the notebook
%matplotlib inline

# read data into a DataFrame
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)

# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

# Modelo de dos parametros - 1 solo feature
lm = smf.ols(formula='sales ~ TV', data=data).fit()

# P Values
print("P Values para TV")
print("----------------")
print(lm.pvalues)
print("Como ninguno es mayor a 5%, no quito ninguno")


# AiC y BIC
print("")
print("AIC Y BIC para TV")
print("-----------------")
print(lm.summary())
```

```python
# Ahora lo mismo con un modelo de 3 parametros
lm = smf.ols(formula='sales ~ TV + radio', data=data).fit()

# P Values
print("P Values para TV + Radio")
print("------------------------")
print(lm.pvalues)
print("Como ninguno es mayor a 5%, no quito ninguno")

# AiC y BIC
print("")
print("AIC Y BIC para TV + Radio")
print("-------------------------")
print(lm.summary())
```

```python
# Ahora lo mismo con un modelo de 4 parametros
lm = smf.ols(formula='sales ~ TV + radio + newspaper', data=data).fit()

# P Values
print("P Values para TV + Radio + Newspaper")
print("-------------------------------------")
print(lm.pvalues)

# AiC y BIC
print("")
print("AIC Y BIC para TV + Radio + Newspaper")
print("-------------------------------------")
print(lm.summary())

print("")
print("Como el P value de newspaper es mayor a 5%, lo quito del modelo")
print("")
print("Resumiendo los IAC y BIC de cada modelo")
print("2 Parametros: AIC=1042 - BIC=1049")
print("3 Parametros: AIC=778.4 - BIC=788.3")
print("4 Parametros: AIC=780.4 - BIC=793.6")
print("Conviene usar el modelo con 3 parametros ya que es el que minimiza AIC y BIC")
```
